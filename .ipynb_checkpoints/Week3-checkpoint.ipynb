## Exercise 1

In this exercise, we will be creating random data ourselves from a given model, and fitting the model to the data. Thereafter the process is repeated a large number of times, and we will look at the statistics concerning the parametres.

---
### Subtask a:

We will assume that between two variables, $x,y$, there is a linear correlation: $y=ax+b$.

Make an $x$ vector, for instance 15 points between -2 and 4 (`np.linspace(-2, 4, 15)`), and generate "true" $y$ values given appropriate values for $a,b$  (pick these yourselves).

Then create the "data" by adding a gaussian distributed number (mean 0) to all the y values. Plot the "true" model as a line along with the noise filled "data" as points with error bars.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
```

```python
a = 2
b = 0.5

def f(x, a, b):
    return a*x + b

x = np.linspace(-2, 4, 15)
y_true = f(x, a, b)

sigma = 1
sigmas = sigma*np.ones_like(x)

y_data = y_true + np.random.normal(0, sigma, 15)


fig, ax = plt.subplots(figsize=(12, 8))

ax.errorbar(x, y_data, sigmas, fmt='.', label='Data')
ax.plot(x, y_true, '-', label='True')
ax.legend();
```

---
### Subtask b:

Make a fit to the data and find estimates for the parametres $a,b$  (${\hat {a}},{\hat {b}}$) as well as their uncertainties $\sigma _{a}, \sigma _{b}$. Refer to Uge_1 for a suggestion of a fitting function that outputs estimates for the parameter errors. Plot the fitted function with the data.

Calculate $\chi^{2}$.

```python
from scipy.optimize import curve_fit

popt, pcov = curve_fit(f, x, y_data, sigma=sigmas)

a_hat, b_hat = popt
a_sigma = np.sqrt(pcov[0,0])
b_sigma = np.sqrt(pcov[1,1])

print(f'a_fit = {a_hat:.3f} +/- {a_sigma:.3f}')
print(f'b_fit = {b_hat:.3f} +/- {b_sigma:.3f}')

def calc_chi2(y_data, y_fit, sigmas):
    return np.sum( (y_data - y_fit)**2/sigmas**2 )

y_fit = f(x, a_hat, b_hat)

chi2 = calc_chi2(y_data, y_fit, sigmas)
print(f'Chi^2 = {chi2:.3f}')


ax.plot(x, y_fit, '-', label='Fit')
ax.legend();

fig
```

### Subtask c:

Repeat the above procedure by generation of random numbers and fitting a large number of times - e.g. $500$. Save only the fit values, the errors on these, as well as $\chi^{2}$ for every iteration.

Compare the standard deviation (p dansk ogs tit kaldet spredning) of the fitting parametres with the estimated errors from the fitting function. Are they approximately equal? Should they be?

Finally, plot all combinations of $\hat{a},\hat{b}$ in a scatter plot. Does the plot reveal anything about the correlation between the estimates for $a$  and $b$?

__Bonus__: _Create histograms of $\hat{a}, \hat{b}, \sigma_{a}, \sigma_{b}$  and look at their distributions_.


```python
N = 500
np.random.deed(45)
MC_results = np.zeros((N, 5))

for i in range(N):
    y_data = y_true + np.random.normal(0, sigma, 15)
    
    popt, pcov = curve_fit(f, x, y_data, sigma=sigmas)

    a_hat, b_hat = popt
    a_sigma = np.sqrt(pcov[0,0])
    b_sigma = np.sqrt(pcov[1,1])

    y_fit = f(x, a_hat, b_hat)
    
    chi2 = calc_chi2(y_data, y_fit, sigmas)

    MC_results[i, :] = a_hat, a_sigma, b_hat, b_sigma, chi2
    
MC_results


print(f'a_MC = {np.mean(MC_results[:, 0]):.3f} +/- {np.std(MC_results[:, 0]):.3f}')
print(f'b_MC = {np.mean(MC_results[:, 2]):.3f} +/- {np.std(MC_results[:, 2]):.3f}')

fig_ab, ax_ab = plt.subplots(figsize=(12, 8))
ax_ab.scatter(MC_results[:, 0], MC_results[:, 2])
ax_ab.set(xlabel='a_MC', ylabel='b_MC');
```

---
### Subtask d:

Make a histogram of the $\chi^{2}$  values. Optionally normalize the histogram.

Plot a $\chi^{2}$ distribution with an appropriate number of degrees of freedom on the same plot. Does everything fit?

```python
fig_chi2, ax_chi2 = plt.subplots(figsize=(12,8))

xaxis = np.linspace(0, 30, 1000)
df = 15-2
chi2_teo = stats.chi2.pdf(xaxis, df)

ax_chi2.hist(MC_results[:, 4], 100, range=(0, 30), histtype='step', lw=1.5, density=True)
ax_chi2.plot(xaxis, chi2_teo, '-', lw=2);
```

## Exercise 2

In this exercise, we will be having a look under the hood of MATLAB's parameter error machine. MATLAB (and almost all other fitting routines) makes a numerical estimate of the errors under the assumption that your function is neat in " $\chi^{2}$-land". This is to be understood in the way that if you look at $\chi^{2}$ as a function of the parametres, the "landscape" is, with decent accuracy, a second degree polynomial in all directions.

If you try to estimate 1 parameter, $p$, by minimizing $\chi^{2}$ in a one dimensional fit, it is true that the best estimate for $p$ is found when $\chi^{2}$ is smallest, and that the $1\sigma$ limits for $p$ is where $\chi^{2}(p)=\chi_{\mathrm{min}}^{2}+1$.

If you have a two-parametre fit, the same way of thinking applies - only the limit is found as $\chi^{2}=\chi _{\mathrm{min}}^{2}+2.3$. The value to be inserted is found in a table of critical $\chi^{2}$-values, or in a work like this one: [[1]](https://stattools.crab.org/Calculators/chiSquareCalculator.htm) (__XXX__ Link in wiki not working). The number of degrees of freedom is equal to the number of parametres you wish to estimate and "propability"/"p value"/"critical value" is set correspondingly to the wanted certainty on the estimate: e.g. set Degrees of Freedom to 2 and Probability to 0,682689 (yes, with a comma).

### Subtask a:

Open the data from the file `Exc2.txt`. The file contains three columns: $x,y,\sigma_{y}$. Plot the data with error bars.

```python
data = np.loadtxt('Exc2.txt', delimiter=',')
x, y, y_sigma = data.T

fig2, ax2 = plt.subplots(figsize=(12, 8))
ax2.errorbar(x, y, y_sigma, fmt='.', label='Data');
```

---
### Subtask b:

We wish to fit the data to a function of the form: $y(x)=a*\mathrm{sinc}((1+b)*x/\pi)$, where $\mathrm{sinc}=\sin(\pi x)/(\pi x)$ is the normalised sinc function.

Define this function, optionally via numpy's built-in `sinc()`, and make a non-linear fit with `curve_fit`. Plot the fitted function together with the data.


```python
def f(x, a, b):
    return a*np.sinc((1+b)*x/np.pi)

popt, pcov = curve_fit(f, x, y, sigma=y_sigma)

a_fit, b_fit = popt
a_std, b_std = np.sqrt(pcov[0,0]), np.sqrt(pcov[1,1])
print(f'a_fit = {a_fit:.3f} +/- {a_std:.3f}')
print(f'b_fit = {b_fit:.3f} +/- {b_std:.3f}')


xaxis = np.linspace(0, 10, 1000)
yaxis = f(xaxis, a_fit, b_fit)

ax2.plot(xaxis, yaxis, '-', label='Fit')

ax2.legend()
fig2
```

### Subtask c:

Now that we have an estimate of the best parametres, we will calculate and plot $\chi^{2}$ for a number of parameter combinations around this optimal point.

1. Begin by making two arrays for $a$ and $b$ in reasonable intervals with enough points (N $\approx$ 100 is fine for this) (e.g. by using `np.linspace`). 

2. You can now loop over all combinations of $a_{i},b_{j}$, and for each combination calculate $\chi^{2}$. The value of each is saved to the array built for exactly that purpose (e.g. `chi2_ab[i, j] = chi2_val` in the loop).

3. Plot the values as a colour coded 2d-plot with for example ax.imshow (note: Long story short: ax.imshow turns everything around in a silly way, so to make things turn intuitively, the array with chi values is transposed: `chi2_ab = chi2_ab.T`. Also remember to set `extent=[xmin, xmax, ymin, ymax]` and `origin='lower'`.
It can be an advantage to plot $\chi^{2}-\chi_{\mathrm{min}}^{2}$ instead of $\chi^{2}$, since this makes everything a bit easier in the next subtask.



```python
def calc_chi2(y_fit, y, sigma):
    return np.sum((y-y_fit)**2/sigma**2)

chi2_fit = calc_chi2(f(x, a_fit, b_fit), y, y_sigma)

Na = 100
Nb = 100
amin, amax = 1.2, 2
bmin, bmax = 0.1, 0.6

a_array = np.linspace(amin, amax, Na)
b_array = np.linspace(bmin, bmax, Nb)

chi2_ab = np.zeros((Na, Nb))

for i, ai in enumerate(a_array):
    for j, bj in enumerate(b_array):
        y_fit = f(x, ai, bj)
        chi2 = calc_chi2(y_fit, y, y_sigma)
        chi2_ab[i, j] = chi2

Z = chi2_ab - chi2_ab.min()
# Notice: have to transpose Z to match up with imshow
Z = Z.T
```

```python
fig, ax = plt.subplots(figsize=(14, 7))

# notice that we are setting the extent and origin keywords
CS = ax.imshow(Z, extent=[amin, amax, bmin, bmax], cmap='viridis_r', origin='lower')
ax.set(xlabel='a', ylabel='b')
fig.colorbar(CS);
```

### Subtask d:

Now we reach the point where we can compare the uncertainties from `curve_fit` with the "true" uncertainty.

1. Draw contour lines on the plot from the last sub task at the values $\chi_{\mathrm{min}}^{2}+1$ and $\chi_{\mathrm {min}}^{2}+2.3$ (i.e. 1 and 2.3 if you have already subtracted $\chi_{\mathrm{min}}^{2}$). For this you can use `ax.contour`.

2. Draw vertical respectively horizontal lines marking the confidence intervals you got from `curve_fit`, i.e. for $a$ , you are to draw lines at $\mu_a+\sigma_a$ and $\mu_a-\sigma_a$. Optionally, add a colorbar.

If you have done everything correctly you should now have a plot showing you which error Scipy (tries to) return(s): The one parameter error - that is, the projection of the contour line at $\Delta \chi^{2}=1$ on the axes. The function handles it pretty well in one direction, and not terribly well in the other, where the error is underestimated.


```python
sigma_levels = [1, 2.3, 6.18, 11.829, 19.33, 28.74, 40, ]
sigma_probs = [sp.special.erf(x/np.sqrt(2)) for x in range(1,9)]
sigma_levels = [stats.chi2.ppf(P, df=2) for P in sigma_probs]
sigma_levels = [1] + sigma_levels


A, B = np.meshgrid(a_array, b_array)

CS = ax.contour(A, B, Z, levels=sigma_levels)
ax.clabel(CS, inline=1, fontsize=10, fmt='%.2f')
ax.set(title= r'Contour plot of $\chi^{2}-\chi_{\mathrm{min}}^{2}$ with $\sigma$ contour levels');

ax.plot(a_fit, b_fit, 'rx')
ax.plot([a_fit-a_std, a_fit-a_std], [bmin, bmax], 'r--')
ax.plot([a_fit+a_std, a_fit+a_std], [bmin, bmax], 'r--')
ax.plot([amin, amax], [b_fit+b_std, b_fit+b_std], 'r--')
ax.plot([amin, amax], [b_fit-b_std, b_fit-b_std], 'r--')

fig
```

## Exercise 3:

Repeat exercise 1, but this time add a small second order term, $y(x)=ax^{2}+bx+c,a\ll 1$, to the y data. You must still fit with the linear function.

It can be useful to look at the residuals, i.e. the data points minus the fit. This can help reveal systematic errors not described by the fit.

What happens to the distribution of the fitting parametres and $\chi^{2}$?

```python
c = 0.2

a = 2
b = 0.5

def f(x, a, b):
    return a*x + b

x = np.linspace(-2, 4, 15)
y_true = a*x + b + c*x**2

sigma = 1
sigmas = sigma*np.ones_like(x)

y_data = y_true + np.random.normal(0, sigma, 15)


fig, ax = plt.subplots(figsize=(12, 8))

ax.errorbar(x, y_data, sigmas, fmt='.', label='Data')
ax.plot(x, y_true, 'o', label='True')
ax.legend();
```

```python

popt, pcov = curve_fit(f, x, y_data, sigma=sigmas)

a_hat, b_hat = popt
a_sigma = np.sqrt(pcov[0,0])
b_sigma = np.sqrt(pcov[1,1])

print(f'a_fit = {a_hat:.3f} +/- {a_sigma:.3f}')
print(f'b_fit = {b_hat:.3f} +/- {b_sigma:.3f}')

def calc_chi2(y_data, y_fit, sigmas):
    return np.sum( (y_data - y_fit)**2/sigmas**2 )

y_fit = f(x, a_hat, b_hat)

chi2 = calc_chi2(y_data, y_fit, sigmas)
print(f'Chi^2 = {chi2:.3f}')


ax.plot(x, y_fit, '-', label='Fit')
ax.legend();

fig
```
